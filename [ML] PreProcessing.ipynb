## 결측치 처리

결측치를 처리하는 방법은 여러가지가 존재함. 크게 보았을 때, 2가지로 보임
- 결측치를 없애는 방법 (없애는 것도 경우를 따지면서 삭제)
- 결측치를 채우는 방법 (평균으로 채우는 등 다양한 방법 존재)

1. Age(나이) : 약 200개의 결측치 존재
   > 이름에 "Mr.", "Ms." 등의 존칭을 붙임.\
   > 존칭의 평균으로 결측치를 채우는게 합당해보임.

2. Cabin(수하물 번호) : 약 700개의 결측치 존재
   > 과연 수하물의 번호가 생존에 영향을 줄 수 있는지 고민.\
   > 영향을 주기 힘들다고 판단 => 변수 자체를 제거해도 괜찮아보임.

### Age 결측치 처리과정

- 이름을 FN(First name 성), LN(Last name 이름), ToH(title of honor 존칭)으로 나누기
- 나누어진 새로운 변수 3개를 맵핑
- 존칭의 평균으로 결측치를 채우기

```python
# 이름에 들어간 ,과 .에 따라  나누기
name_df = titanic_csv["Name"].str.split("[,.]", 2, expand=True)
name_df.columns = ["FN", "ToH", "LN"]  # 레이블 

# 생성된 변수의 데이터들의 공백 제거
for i in range(2) : 
    name_df[name_df.columns[i]] = name_df[name_df.columns[i]].str.strip()

# 잘 처리되었는지 확인
name_df.value_counts()
```

```python
# 존칭만 우선 확인
name_df["ToH"].value_counts()
```

```python
# 새로만든 변수를 타이타닉 데이터에 이어 붙이기
titanic_df = pd.concat((titanic_csv,name_df), axis=1)
titanic_df
```

```python
# 특정할 수 있는 존칭을 제외한 나머지 존칭들을 하나로 묶는 상황 
titanic_df.loc[
  ~((titanic_df["ToH"]=="Mr")|
    (titanic_df["ToH"]=="Miss")|
    (titanic_df["ToH"]=="Mrs")|
    (titanic_df["ToH"]=="Marster")|
    (titanic_df["ToH"]=="Dr")|
    (titanic_df["ToH"]=="Rev")    ),
    "ToH"] = "others"

titanic_df['ToH'].value_counts()
```

```python
# 칭호에 따른 나이 평균 구하기
toh_age_mean = titanic_df[["ToH","Age"]].groupby("ToH").mean().reset_index()
toh_age_mean.columns = ["ToH","toh_Age"]
toh_age_mean
```

```python
# 나이평균 존칭에 따라 합치기
titanic_df = pd.merge(titanic_df, toh_age_mean, on = "ToH", how = "left")

# 비어있는 나이 결측 채우기
titanic_df.loc[(titanic_df["Age"].isnull()), "Age"] = titanic_df["toh_Age"]

# 필요없는 존칭 평균나이 없애기
titanic_df = titanic_df.drop(["toh_Age"], axis=1)

titanic_df
```

## 변수 제거(Cabin 변수 포함)

- 없애면서 필요없을거 같은 변수 모두 제거

```python
# 제거 후 확인
titanic_df = titanic_df.drop(["PassengerId", "Name", "Ticket", "Cabin", "FN", "LN"], axis=1)
titanic_df
```

## 새로운 변수 생성

남아 있는 변수 목록
1. Survived : 생존 여부 (0 = 사망, 1 = 생존)
2. Pclass : 티켓 클래스 (1 = 1등석, 2 = 2등석, 3 = 3등석)
3. Sex : 성별
4. Age : 나이
5. SibSp : 함께 탑승한 자녀 / 배우자의 수
6. Parch : 함께 탑승한 부모님 / 아이들의 수
7. Fare : 탑승 요금
8. Embarked : 선착장 (C = Cherbourg, Q = Queenstown, S = Southampton)
9. FN : 성
10. ToH : 존칭
11. LN : 이름

- 의문이 드는 변수 sibsp, parch => 두 변수 모두 같이 탑승한 사람을 물어보고 있음. 하지만 둘 다 자녀의 수를 물어봄
- 3가지 방법을 생각함(두 변수 중 수가 큰 변수를 사용, 두 변수를 합치고 2로 나누어 사용, 아니면 사용 안함)


```python
# 두 변수를 합치고 2로 나누어 사용 하는 방법으로 결정
# Family(Family) = (sibsp + parch) / 2
family = (titanic_df["SibSp"]+titanic_df["Parch"])/2
family = family.astype("int64")
family
```

```python
titanic_df = pd.concat((titanic_df,family), axis=1)
titanic_df.rename(columns={0:"Family"}, inplace = True)  #inplace = True(저장함)
titanic_df
```

```python
# 제거 후 확인
titanic_df["Age"] = titanic_df["Age"].astype("int")
titanic_df = titanic_df.drop(["SibSp", "Parch"], axis=1)
titanic_df
```
